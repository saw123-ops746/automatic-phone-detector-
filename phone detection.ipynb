```python
import cv2
import os
import time 
import winsound
import threading
from ultralytics import YOLO

import tkinter as tk
from tkinter import Label, Button

# Load YOLO model
model = YOLO("yolov8n.pt")

# Create screenshots folder
if not os.path.exists("screenshots"):
    os.makedirs("screenshots")

# Global variables
phone_detect_count = 0
alarm_on = False
running = False
cap = None

def play_alarm():
    # Use winsound for Windows, or replace with playsound if you want cross-platform
    winsound.Beep(1000, 500)  # Beep at 1000 Hz for 500 ms

def start_detection():
    global cap, running, phone_detect_count, alarm_on
    cap = cv2.VideoCapture(0)
    running = True
    while running:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame from camera.")
            break
        results = model(frame)

        phone_detected = False

        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            confidence = float(box.conf[0])

            # Detect only cell phone with confidence > 60%
            if model.names[cls_id] == "cell phone" and confidence > 0.6:
                phone_detected = True
                phone_detect_count += 1
                x1, y1, x2, y2 = map(int, box.xyxy[0])

                # Draw ONLY phone box
                cv2.rectangle(frame, (x1, y1), (x2, y2), (2, 2, 120), 2)
                cv2.putText(frame, f"PHONE {confidence:.2f}",
                            (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX,
                            0.8,
                            (0, 0, 255),
                            2)

        if phone_detected:
            print(" Phone Detected!")
            if not alarm_on:
                alarm_on = True
                threading.Thread(target=play_alarm).start()
        else:
            alarm_on = False

        # Update detection count label
        count_label.config(text=f"Detections: {phone_detect_count}")

        # Show original frame (not plot())
        cv2.imshow("Mobile Phone Detector", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    if cap is not None:
        cap.release()
    
    alarm_on = False
    print("Detection stopped.")

# ================= GUI ==================

root = tk.Tk()
root.title("Classroom Mobile Detector")
root.geometry("400x200")

title_label = Label(root, text=" Classroom Phone Detector", font=("Arial", 16))  # Fixed font size
title_label.pack(pady=10)

count_label = Label(root, text="Detections: 0", font=("Arial", 14))
count_label.pack(pady=10)

start_button = Button(root, text="Start Detection", command=lambda: threading.Thread(target=start_detection).start())
start_button.pack(pady=5)

def stop_detection():
    global running, cap
    running = False
    if cap is not None:
        try:
            cap.release()
        except Exception:
            pass
    print("Detection stopped.")

stop_button = Button(root, text="Stop Detection", command= stop_detection)
stop_button.pack(pady=5)

root.mainloop()   
```
